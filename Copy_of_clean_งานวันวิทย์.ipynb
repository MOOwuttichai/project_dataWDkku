```python
# prompt: ช่วย concat ข้อมูล SC01 ถึง SC13
# Concatenate all DataFrames into one
combined_df = pd.concat([SC01, SC02, SC03, SC04, SC05, SC06, SC07, SC08, SC09, SC10, SC11, SC12, SC13], ignore_index=True)
# Now you have a single DataFrame with all the data
combined_df
```

```python
# prompt: ในข้อมูล combined_df เอา 2 column สุดท้ายออก

# Drop the last two columns
combined_df = combined_df.iloc[:, :-2]
```

```python
# prompt: ในข้อมูล combined_df ค่า NaN ใน column V2 เเก้เป็น 9
combined_df['V2'] = combined_df['V2'].fillna(9)
combined_df['V8_11'] = combined_df['V8_11'].fillna(9)
combined_df['V11'] = combined_df['V11'].fillna(99)
```

```python
combined_df[combined_df['V8_1']==99]
```

```python
# prompt: ช่วยเเก้ค่า 9 เป็น 99 ใน combined_df ยกเว้นcolumn V9_1 ,V9_2

# Replace 9 with 99 in all columns except 'V9_1' and 'V9_2'
for col in combined_df.columns:
  if col not in ['หมายเลขแบบสอบถาม','V9_1', 'V9_2','V1']:
    combined_df[col] = combined_df[col].replace(9, 99)
```

```python
# prompt: ในข้อมูล combined_df ให้ลบ rowที่มีค่า 99 ในcolumn V2 ,V3 ,V4 ,V5 ,V10_1 ,V10_2 ,V10_3 ,V10_4 ,V10_5 ,V10_6 ,V11

# Create a boolean mask to identify rows with 99 in the specified columns
mask = (combined_df['V2'] == 99) | (combined_df['V3'] == 99) | (combined_df['V4'] == 99) | \
        (combined_df['V5'] == 99) | (combined_df['V10_1'] == 99) | (combined_df['V10_2'] == 99) | \
        (combined_df['V10_3'] == 99) | (combined_df['V10_4'] == 99) | (combined_df['V10_5'] == 99) | \
        (combined_df['V10_6'] == 99) | (combined_df['V11'] == 99)

# Invert the mask to keep rows that do NOT have 99 in the specified columns
combined_df = combined_df[~mask]
```
