```python
# prompt: จินีผมต้องการใช้ python sentence transformers สำหรับ clustering เเล้วมีการทดสอบประสิทธิภาพด้วยช่วยหน่อยสิ

!pip install sentence-transformers

from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans

# Load the SentenceTransformer model
model = SentenceTransformer('all-mpnet-base-v2')

# Encode the sentences
sentences = ['This is a sentence.', 'This is another sentence.', 'This is a third sentence.']
embeddings = model.encode(sentences)

# Cluster the sentences using KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(embeddings)

# Print the cluster labels
print(kmeans.labels_)

# Evaluate the clustering performance
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(embeddings, kmeans.labels_)
print("Silhouette score:", silhouette_avg)
```
